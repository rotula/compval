```python
import sqlite3
import re
```

```python
sql = 'SELECT * FROM formdesc'
conn = sqlite3.connect('compval')
```

```python
tokenlist = set()
for fname, hd_leaves, hd_pages, hd_type, hd_script, div_rend, div_rendition, div_type in conn.execute(sql):
    s = ' '.join([div_rend, div_rendition, div_type]).strip()
    s = re.compile(r'[;,.]+').sub(' ', s)
    s = re.compile(r'(\d+)(Bl|S)').sub('\\1 \\2', s)
    # print(f'{fname}: {s}')
    leaves = []
    pages = []
    typs = []
    scripts = []
    tokenizer = (t for t in re.compile(r'\s+').split(s))
    for t in tokenizer:
        if t.lower().startswith('x'):
            num = t
        elif t.isdigit():
            num = t
        elif t == 'Bl':
            if not num:
                print(f'{fname}: "{s}", no leaf number given')
            leaves.append(f'{num} Bl.')
            num = None
        elif t == 'S':
            if not num:
                print(f'{fname}: "{s}", no page number given')
            pages.append(f'{num} S.')
            num = None
        elif t in ['lt', 'tg', 'pc']:
            typs.append(t)
        elif t in ['ms', 'ts']:
            scripts.append(t)
        elif t == '?':
            pass
        else:
            print(f'{fname}: "{s}", unknown token "{t}"')

    if num and num.isdigit():
        print(f'{fname}: "{s}", ends with number')
    if len(leaves) > 1:
        print(f'{fname}: "{s}", too many leaves')
    if len(pages) > 1:
        print(f'{fname}: "{s}", too many pages')
    if len(typs) > 1:
        print(f'{fname}: "{s}", too many types')
    if len(scripts) > 1:
        print(f'{fname}: "{s}", too many scripts')

    try:
        if hd_leaves != leaves[0]:
            print(f'{fname}: "{s}": "{hd_leaves}" != "{leaves[0]}"')
    except (IndexError, ) as e:
        print(f'{fname}: "{s}", but "{hd_leaves}" in SQL')

    try:
        if hd_pages != pages[0]:
            print(f'{fname}: "{s}": "{hd_pages}" != "{pages[0]}"')
    except (IndexError, ) as e:
        print(f'{fname}: "{s}", but "{hd_pages}" in SQL')

    try:
        if hd_type != typs[0]:
            print(f'{fname}: "{s}": "{hd_type}" != "{typs[0]}"')
    except (IndexError, ) as e:
        print(f'{fname}: "{s}", but "{hd_type}" in SQL')

    try:
        if hd_script != scripts[0]:
            print(f'{fname}: "{s}": "{hd_script}" != "{scripts[0]}"')
    except (IndexError, ) as e:
        print(f'{fname}: "{s}", but "{hd_script}" in SQL')




```
